# 2. How are in-demand skills trending for Data Analysts?

```python
# Importing Libraries & Dataset
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\Data_Science\Projects\Python_Data_Analyst_Job_Market_Analysis\Dataset\data_jobs.csv')
```

```python
# Cleaning Data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Filter: Data Analyst Jobs in US
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
df_DA_US.head(3)
```

```python
# Extract Month Information
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
df_DA_US['job_posted_month_no']
```

```python
# Explode Skills
df_DA_US_explode = df_DA_US.explode('job_skills')
df_DA_US_explode['job_skills']
```

```python
# Pivot Table of Skills per Month
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Add totals row, sort columns by popularity, then drop totals
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot= df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
# Normalize to Percentages
DA_totals = df_DA_US.groupby('job_posted_month_no').size()

DA_totals
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

df_DA_US_percent
```

```python
df_DA_US['job_posted_month_no'].dtype
```

```python
# Map Month Numbers to Names
import calendar

# Reformat month into abbreviated names
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.strftime('%b')

month_order = list(calendar.month_abbr)[1:]  # ['Jan', 'Feb', ..., 'Dec']
df_DA_US['job_posted_month_no'] = pd.Categorical(
    df_DA_US['job_posted_month_no'],
    categories=month_order,
    ordered=True
)

df_DA_US_pivot.index = df_DA_US_pivot.index.map(lambda i: calendar.month_abbr[i])
df_DA_US_percent.index = df_DA_US_percent.index.map(lambda i: calendar.month_abbr[i])
```

```python
df_DA_US_percent
```

```python
# Select Top 5 Skills
df_plot = df_DA_US_percent.iloc[:, :5]

df_plot
```

```python
# Plot: Skill Trends
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Postings')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
